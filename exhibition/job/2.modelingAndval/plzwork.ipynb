```python
import pandas as pd
df1 = pd.read_csv(r'C:\Users\brian\Desktop\JUNSOO\exhibition\data\temp.csv' , encoding ='cp949')
df2 = pd.read_csv(r'C:\Users\brian\Desktop\JUNSOO\exhibition\data\final_data.csv' , encoding = "euc-kr")
df1 = df1.loc[df1['전체학생수']!= 0 , :]
```

```python
print(df1.columns.tolist())

col_use = ['key' , '전체학생수' , '상담실적(외부상담전문가)'
    , '연간1인당 보건실이용건수' , '교수학습공간 일반교실'
    , '교수학습공간 특별교실' 
    , '교수학습공간 수준별교실' ,'참여학생수-동아리?학생자치활동' , '참여학생수-또래활동' 
           , '참여학생수-교육주간 활동' , '참여학생수-기타 학교폭력 예방활동'
           , '일반직계' , '별정직계(2018년이후)' , '기타직계(2019년이후)' ,'전문상담교사(계)' ,
           '수석교사(계)' , '체력 1등급 (%)' , '교육청금액' , '학교교육시설 보수확충금액', '수목시설및 재산금액'
            ,'도서및 물품금액','금전및 유가증권금액', '자치단체금액' ,'상담실적(내부상담전문가)'
           ]
df1 = df1[col_use]
```

```python
df1.info()
df1_copy = df1.copy()
```

```python
# 수치형 데이터의 범위를 바탕으로 구간을 나누어 레이블링하기
condition = ((df1['참여학생수-동아리?학생자치활동'] >=1) & (df1['참여학생수-동아리?학생자치활동'] < 53))
df1.loc[condition , '동아리_class' ] = 1
condition2 = (df1['참여학생수-동아리?학생자치활동'] >= 53)
df1.loc[condition2 , '동아리_class'] = 2
df1.loc[df1['참여학생수-동아리?학생자치활동'] == 0 ,'동아리_class'] = 0
```

```python
# 1) 학교평가 공통지표
# 수준별 교육과정 , 학생이해 노력, 전문적 공동체 구축
# 교육활동지원을 위한 행정, 재정적 지원
# 수석교사의 수
df1['t1'] = df1['전체학생수']
df1['t2'] = df1['수석교사(계)']

# 2) 목적 사업비
# 기초자치단체 이전수입

df1['t3'] = df1['연간1인당 보건실이용건수']
df1['t4'] = df1['별정직계(2018년이후)'] + df1['기타직계(2019년이후)']
df1['t5'] = df1['동아리_class']


# 3) 진로성숙도
# 진로체험경험 , 학교 진로지원 , 학업성취도 ,부모와 자녀간의 의사소통
df1['t6'] = df1['상담실적(외부상담전문가)'] + df1['상담실적(내부상담전문가)']
df1['t7'] = df1['전문상담교사(계)']
# 
```

```python
df1_pay_score = df1.loc[:,'t1':'t7']
df1_pay_score.head().T

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df1_pay_score["y2"] = scaler.fit_transform(df1_pay_score).sum(axis = 1).tolist()
```

```python
df1_key = pd.DataFrame(df1['key'])
df1_pay_score = pd.concat([df1_pay_score , df1_key] , axis = 1)
df1_pay_score
```
